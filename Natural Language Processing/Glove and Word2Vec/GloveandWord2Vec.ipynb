```python
def load_glove_vectors(tmp_file):
    return KeyedVectors.load_word2vec_format(tmp_file, binary=False)

word_vectors = load_glove_vectors(tmp_file)

# Perform analogy
def analogy(word_a, word_b, word_c, word_vectors):
    vec = word_vectors[word_b] - word_vectors[word_a] + word_vectors[word_c]
    return word_vectors.similar_by_vector(vec, topn=1)[0][0]

# Perform analogy
result = analogy('king', 'man', 'woman', word_vectors)
print(result)
```

```python
def display_closestwords_tsnescatterplot(model, dim, words):
    
    arr = np.empty((0,dim), dtype='f')
    word_labels = words

    # get close words
    #close_words = [model.similar_by_word(word) for word in words]
    
    # add the vector for each of the closest words to the array
    close_words=[]
    for word in words:
        arr = np.append(arr, np.array([model[word]]), axis=0)
        close_words +=model.similar_by_word(word)
        
    for wrd_score in close_words:
        wrd_vector = model[wrd_score[0]]
        word_labels.append(wrd_score[0])
        arr = np.append(arr, np.array([wrd_vector]), axis=0)
        
    # find tsne coords for 2 dimensions
    tsne = TSNE(n_components=2, random_state=0)
    #np.set_printoptions(suppress=True)
    Y = tsne.fit_transform(arr)

    x_coords = Y[:, 0]
    y_coords = Y[:, 1]
    # display scatter plot
    plt.scatter(x_coords, y_coords)

    for label, x, y in zip(word_labels, x_coords, y_coords):
        plt.annotate(label, xy=(x, y), xytext=(0, 0), textcoords='offset points')
    plt.xlim(x_coords.min()+0.00005, x_coords.max()+0.00005)
    plt.ylim(y_coords.min()+0.00005, y_coords.max()+0.00005)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE

def tsne_plot(model, words):
    "Creates and TSNE model and plots it"
    vectors = np.array([model[word] for word in words])
    labels = words
    
    tsne_model = TSNE(perplexity=40, n_components=2, init='pca', n_iter=2500, random_state=23)
    new_values = tsne_model.fit_transform(vectors)

    x = []
    y = []
    for value in new_values:
        x.append(value[0])
        y.append(value[1])
        
    plt.figure(figsize=(16, 16)) 
    for i in range(len(x)):
        plt.scatter(x[i],y[i])
        plt.annotate(labels[i],
                     xy=(x[i], y[i]),
                     xytext=(5, 2),
                     textcoords='offset points',
                     ha='right',
                     va='bottom')
    plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Function to plot clusters
def plot_clusters(embeddings, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(embeddings[:, 0], embeddings[:, 1], c=labels, cmap='viridis', alpha=0.5)
    plt.title(title)
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.colorbar(label='Cluster')
    plt.show()

# Extract embeddings for the last 100 words
words = model.index_to_key[-100:]
embeddings = [model[word] for word in words]
embeddings_array = np.array(embeddings)

# Perform K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(embeddings_array)

# Perform Gaussian Mixture Model clustering
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(embeddings_array)

# Plot the clusters for Word2Vec embeddings
plot_clusters(embeddings_array, kmeans_labels, 'K-Means Clustering for Word2Vec Embeddings')
plot_clusters(embeddings_array, gmm_labels, 'GMM Clustering for Word2Vec Embeddings')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.manifold import TSNE
from gensim.models import KeyedVectors

# Load your AutoKeyedVectors model
#path = '/Users/mac/Downloads/GoogleNews-vectors-negative300.bin.gz'
#model = KeyedVectors.load_word2vec_format(path, binary=True, limit=20000)


def tsne_plot(model, words, title, figsize=(10, 8), annotate=True, alpha=0.5, perplexity=30):
    # Extract embeddings for the provided words
    embeddings = [model[word] for word in words]
    embeddings_array = np.array(embeddings)
    
    # Initialize t-SNE model
    tsne_model = TSNE(n_components=2, random_state=42, perplexity=perplexity)

    # Fit and transform the data
    tsne_embeddings = tsne_model.fit_transform(embeddings_array)

    # Plot the t-SNE embeddings
    plt.figure(figsize=figsize)
    plt.scatter(tsne_embeddings[:, 0], tsne_embeddings[:, 1], alpha=alpha)
    
    if annotate:
        for i, word in enumerate(words):
            plt.annotate(word, (tsne_embeddings[i, 0], tsne_embeddings[i, 1]))
    
    plt.title(title)
    plt.xlabel('t-SNE Dimension 1')
    plt.ylabel('t-SNE Dimension 2')
    plt.grid(True)  # Adding grid
    plt.tight_layout()  # Adjust layout
    plt.show()
```
